```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3,random_state=42)
df["prediction"]=km.fit_predict(df)
```

```python
df.head()
```

```python
sns.pairplot(df,hue="prediction")
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(df, km.labels_)
print(score)
```

```python
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df)
    print(k, silhouette_score(df, labels))
```

```python
abc=[]
for i in range(2,14):
    knn=KMeans(n_clusters=i)
    knn.fit_predict(df)
    abc.append(knn.inertia_)

plt.plot(range(2,14),abc,marker="o")
plt.xticks([i for i in range(2,14)])
plt.grid(axis="x")
plt.xlabel("no of clusters")
plt.ylabel("inertia")
plt.show()
```
