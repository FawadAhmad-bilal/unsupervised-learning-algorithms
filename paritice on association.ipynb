```python
df=df.drop("CustomerID",axis=1)
```

```python
df=df.dropna(subset=["Description"])
```

```python
basket=df.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().fillna(0)
```

```python
encode_basket=(basket>0).astype(int)
encode_bool=encode_basket.astype(bool)
```

```python
encode_bool
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent=apriori(encode_bool,min_support=0.04,use_colnames=True,max_len=2).sort_values(by=["support"])
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rule=association_rules(frequent,metric="lift",min_threshold=1.0)
rule=rule.sort_values("lift",ascending=False)
print(rule[["antecedents", "consequents", "support", "confidence", "lift"]].head(10))
```
