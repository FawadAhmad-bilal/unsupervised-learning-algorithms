```python
df=df.dropna(subset=["Description"])
```

```python
df.shape
```

```python
basket=df.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().fillna(0)
```

```python
basket_encode=(basket>0).astype(int)
basket_bool=basket_encode.astype(bool)
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
frequent_itemsets=fpgrowth(basket_bool,min_support=0.03,use_colnames=True,max_len=3).sort_values(by=["support"])
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.5
)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules
```

```python
strong_rules = rules[rules['lift'] > 3]
strong_rules
```
